```python
import numpy as np
import pandas as pd

#zad1 
data = {'War': [1, 2, 3, 4, 5, 6], 'Prawd': [1/6, 1/6, 1/6, 1/6, 1/6, 1/6]}
df = pd.DataFrame(data)

mean = (df['War'] * df['Prawd']).sum()
variance = (((df['War'] - mean) ** 2) * df['Prawd']).sum()
std_dev = np.sqrt(variance)
median = df.loc[df['Prawd'].cumsum() >= 0.5, 'War'].min()
print('Średnia: ', round(mean, 2))
print('Wariancja: ', round(variance, 2))
print('Odchylenie : ', round(std_dev, 2))
print('Mediana: ', round(median, 2))
```

```python
#zad2
import numpy as np


bernoulli_samples = np.random.binomial(n=1, p=0.5, size=100)
print(bernoulli_samples)

binomial_samples = np.random.binomial(n=10, p=0.5, size=100)
print(binomial_samples)

poisson_samples = np.random.poisson(lam=5, size=100)
print(poisson_samples)
```

```python
#zad
import numpy as np

print(' Bernoulliego:')
print('Średnia: ', np.mean(bernoulli_samples))
print('Wariancja: ', np.var(bernoulli_samples))

print(' Dwumianowy:')
print('Średnia: ', np.mean(binomial_samples))
print('Wariancja: ', np.var(binomial_samples))

print(' Poissonao:')
print('Średnia: ', np.mean(poisson_samples))
print('Wariancja: ', np.var(poisson_samples))
```

```python
#zad4
import matplotlib.pyplot as plt

plt.subplot(131)
plt.hist(bernoulli_samples, bins=[0, 0.5, 1], density=True, align='left', rwidth=0.8)
plt.title('Rozkład Bernoulliego')

plt.subplot(132)
plt.hist(binomial_samples, bins=range(11), density=True, align='left', rwidth=0.8)
plt.title('Rozkład Dwumianowy')

plt.subplot(133)
plt.hist(poisson_samples, bins=range(11), density=True, align='left', rwidth=0.8)
plt.title('Rozkład Poissona')
```

```python
#zad 5
from scipy.stats import binom

n = 20
p = 0.4

probabilities = []
for k in range(n+1):
    probability = binom.pmf(k, n, p)
    probabilities.append(probability)
    print(f'Prawdopodobieństwo dla k = {k} : {probability}')

sum_prob = sum(probabilities)
print('Suma prawdopodobieństw: ', round(sum_prob, 2))
```

```python
#zad 6
import numpy as np

data = np.random.normal(loc=0, scale=2, size=100)

mean = np.mean(data)
print('Średnia: ', mean)

variance = np.var(data)
print('Wariancja: ', variance)

std_dev = np.std(data)
print('Odchylenie : ', std_dev)

median = np.median(data)
print('Mediana: ', median)

print()


data = np.random.normal(loc=0, scale=2, size=200)

mean = np.mean(data)
print('Średnia: ', mean)

variance = np.var(data)
print('Wariancja: ', variance)

std_dev = np.std(data)
print('Odchylenie standardowe: ', std_dev)

median = np.median(data)
print('Mediana: ', median)
```

```python
import numpy as np
import matplotlib.pyplot as plt

data_norm = np.random.normal(loc=1, scale=2, size=1000)
data_std = np.random.standard_normal(size=1000)

x = np.linspace(-5, 5, 1000)
y = 1/(0.5*np.sqrt(2*np.pi)) * np.exp(- (x - (-1))**2/(2*0.5**2))

plt.subplot(131)
plt.hist(data_norm, bins=30, density=True, color='blue', alpha=0.6)
plt.title('Rozkład normalny')

plt.subplot(132)
plt.hist(data_std, bins=30, density=True, color='green', alpha=0.6)
plt.title('Rozkład standardowy')

plt.subplot(133)
plt.plot(x, y, color='red')
plt.title('Gęstość rozkładu normalnego')

plt.tight_layout()
plt.show()
```
